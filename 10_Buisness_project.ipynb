```python
import numpy as np
import matplotlib.pyplot as plt
sales_data = np.array([
    [1,15000,20000,30000,40000],    #burger king
    [2,50000,40000,34000,90000],    #kfc
    [3,34000,54000,65000,87000],    #pizza hut
    [4,54000,24000,35000,67000],    #Mc'D
    [5,33000,14000,15000,47000],    #domino's
])
print("--------------Zomato Sales Analyze--------------")
```

```python
#total sales
yearly_sales = np.sum(sales_data[:, 1:],axis=0)
total_sales = np.sum(yearly_sales[0:])
print("sales yearly wise :",yearly_sales)
print("Total sales: ₹",total_sales)

#minimum sales per restaurant
minimum = np.min(sales_data[:,1:],axis=1)
print("Minimum sales of restaurant:",minimum)

#maximum sales of restaurant
maximum = np.max(sales_data[:,1:],axis=1)
maximum1 = np.max(sales_data[:,1:],axis=0)
print("Maximum sales of restaurant:",maximum)
print("Maximum sales of year 2021 , 2022 , 2023 , 2024:",maximum1)

#Average sales 
avg_sales = np.mean(sales_data[:,1:],axis=1)
avg_sales_year = np.mean(sales_data[:,1:],axis=0)
print("average sales of restaurant:",avg_sales)
print("average sales per year:",avg_sales_year)

#cumilative sales
cum = np.cumsum(sales_data[:,1:],axis=1)
print("cumilative sales :\n",cum)
```

--------------Zomato Sales Analyze- Graph-------------

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cum,axis=1))
plt.title("average of cumilative sales")
plt.grid=(True)
plt.xlabel("year")
plt.ylabel("sales")
plt.show()
```

```python
years = ['2021', '2022', '2023', '2024']
brands = ['Burger King', 'KFC', 'Pizza Hut', "Mc'D", "Domino's"]


for i in range(len(sales_data)):
    plt.bar(years, sales_data[i, 1:], label=brands[i])  # skip the brand ID

plt.title('Quarterly Sales of Fast Food Chains')
plt.xlabel('Quarter')
plt.ylabel('Sales in INR')

plt.legend()
plt.show()
```

```python

plt.pie(avg_sales, labels=brands, autopct='%1.1f%%', startangle=100)
plt.title("Average Sales of restaurant")
plt.axis("equal")  
plt.show()
```
