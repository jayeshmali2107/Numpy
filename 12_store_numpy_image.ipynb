```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python

try:
    img = Image.open('numpy-logo.png').convert('RGB')
    logo = np.array(img)

    # Create dark  version
    dark_logo = 255 - logo

    # Save the dark logo
    dark_img = Image.fromarray(dark_logo)
    dark_img.save('dark_numpy_logo.png')

    # Display both logos side by side
    plt.figure(figsize=(10, 5))

    plt.subplot(121)
    plt.imshow(logo)
    plt.title("Numpy logo")
    plt.axis('off')  

    plt.subplot(122)
    plt.imshow(dark_logo)
    plt.title("Numpy Dark logo")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

except FileNotFoundError:
    print("❌ Error: 'numpy-logo.png' not found.")
```
